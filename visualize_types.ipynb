```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV file
file_path = '15_000_preprocessed_train.csv'
df = pd.read_csv(file_path)

# Define type mapping
type_mapping = {
    'unreliable': 'fake',
    'fake': 'fake',
    'conspiracy': 'fake',
    'bias': 'fake',
    'hate': 'fake',
    'junksci': 'fake',
    'clickbait': 'reliable',
    'reliable': 'reliable',
    'political': 'reliable',
    'satire': 'fake',
    'state': 'fake',
    'rumor': 'fake',
}

# Apply mapping to the dataset
df['mapped_type'] = df['type'].map(type_mapping)

# Count number of articles in each category
count_articles = df['mapped_type'].value_counts()
print(count_articles)

# Plot distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="mapped_type", hue="mapped_type", palette="coolwarm", legend=False)
plt.title("Distribution of Types Based on Mapping")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

```
